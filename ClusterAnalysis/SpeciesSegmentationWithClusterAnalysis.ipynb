# Species Segmentation with Cluster Analysis

The Iris flower dataset.
Start by creating 2 clusters. Then standardize the data and try again. Does it make a difference?
Use the Elbow rule to determine how many clusters are there.


## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'iris_dataset.csv'</i>.

```python
raw_data = pd.read_csv('iris_dataset.csv')
raw_data
data = raw_data.copy()
data
```

## Plot the data

For this exercise, try to cluster the iris flowers by the shape of their sepal. 

<i> Use the 'sepal_length' and 'sepal_width' variables.</i> 

```python
plt.scatter(data['sepal_length'],data['sepal_width'] ) 
plt.ylabel('Sepal width')
plt.show()
```

# Clustering (unscaled data)

Separate the original data into 2 clusters.

```python
x = data.copy()
x
```

```python
kmeansUnscaledData = KMeans(2)
kmeansUnscaledData.fit(x)
```

```python
df = data.copy()
df['Cl_pred_unscaled']=kmeansUnscaledData.fit_predict(x)
```

```python
df
```

```python
plt.scatter(df['sepal_length'],df['sepal_width'],c=df['Cl_pred_unscaled'], cmap='rainbow' )
plt.title('Clusters using unscaled data')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

# Standardize the variables

Import and use the <i> method </i> function from sklearn to standardize the data. 

```python
from sklearn import preprocessing
```

```python
x_forscaling = x.copy()
x_forscaling
x_scaled = preprocessing.scale(x_forscaling)
x_scaled
```

# Clustering (scaled data)

```python
kmeansScaledData = KMeans(2)
kmeansScaledData.fit(x_scaled)
```

```python
df['Cl_pred_Scaled']=kmeansScaledData.fit_predict(x_scaled)
```

```python
df
```

```python
plt.scatter(df['sepal_length'],df['sepal_width'],c=df['Cl_pred_Scaled'], cmap='rainbow' )
plt.title('Clusters with Scaled Data')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

```python
fig,axs = plt.subplots(1,2,figsize=(10,5))
axs[0].scatter(df['sepal_length'],df['sepal_width'],c=df['Cl_pred_unscaled'], cmap='rainbow' )
axs[0].set_title('Clusters with Unscaled Data')
axs[0].set_xlabel('Sepal length')
axs[0].set_ylabel('Sepal width')


axs[1].scatter(df['sepal_length'],df['sepal_width'],c=df['Cl_pred_Scaled'], cmap='rainbow' )
axs[1].set_title('Clusters with Scaled Data')
axs[1].set_xlabel('Sepal length')
axs[1].set_ylabel('Sepal width')

plt.tight_layout()
plt.show()
```

## Take Advantage of the Elbow Method

### WCSS

```python
wcss = []
cl_num = range(1,10)
for i in cl_num:
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    cl_wcss = kmeans.inertia_
    wcss.append(cl_wcss)
wcss    
```

### The Elbow Method

```python
len(cl_num)
```

```python
len(wcss)
```

```python
plt.plot(cl_num, wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

How many clusters are there? <br>
The Elbow Method suggests we try 2, 3, or 5 clusters. We have already tried 2 clusters. We will proceed with 3 and 5 clusters.

# 3 Clusters

```python
kmeansScaledData = KMeans(3)
kmeansScaledData.fit(x_scaled)
df['Cl_03_pred_Scaled']=kmeansScaledData.fit_predict(x_scaled)
plt.scatter(df['sepal_length'],df['sepal_width'],c=df['Cl_03_pred_Scaled'], cmap='rainbow' )
plt.title("3 Clusters")
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

# 5 Clusters

```python
kmeansScaledData = KMeans(5)
kmeansScaledData.fit(x_scaled)
df['Cl_05_pred_Scaled']=kmeansScaledData.fit_predict(x_scaled)
plt.scatter(df['sepal_length'],df['sepal_width'],c=df['Cl_05_pred_Scaled'], cmap='rainbow' )
plt.title("5 Clusters")
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

# Compare our solutions to the original iris dataset
The original (full) iris data is located in <i>iris_with_answers.csv</i>. Load the csv, plot the data and compare it with your solution. 

Obviously there are only 3 types, because that's the original (truthful) iris dataset.

The 2-cluster solution seemed good, but in real life the iris dataset has 3 SPECIES (a 3-cluster solution). Therefore, clustering cannot be trusted at all times. Sometimes it seems like x clusters are a good solution, but in real life, there are more (or less).

```python
og_iris_df = pd.read_csv('iris_with_answers.csv')
og_iris_df
```

```python
data = og_iris_df.copy()
data
data['species'].unique()
```

```python
data['species']=data['species'].map({'setosa':0,'versicolor':1,'virginica':2})
data
```

```python
plt.scatter(data['sepal_length'],data['sepal_width'],c=data['species'],cmap='rainbow')
plt.title('Original Iris Data - sepal attributes')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

```python
plt.scatter(data['petal_length'],data['petal_width'],c=data['species'],cmap='rainbow')
plt.title('Original Iris Data - petal attributes')
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.show()
```

```python
kmeansScaledData = KMeans(3)
kmeansScaledData.fit(x_scaled)
df['Cl_03_pred_Scaled_Petals']=kmeansScaledData.fit_predict(x_scaled)
plt.scatter(df['petal_length'],df['petal_width'],c=df['Cl_03_pred_Scaled_Petals'], cmap='rainbow' )
plt.title("3 Clusters on Petal Attributes")
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.show()
```

```python
fig,axs = plt.subplots(1,2,figsize=(10,5))

axs[0].scatter(data['petal_length'],data['petal_width'],c=data['species'],cmap='rainbow')
axs[0].set_title('Original Iris Data - petal attributes | ground truth')
axs[0].set_xlabel('Petal length')
axs[0].set_ylabel('Petal width')

axs[1].scatter(df['petal_length'],df['petal_width'],c=df['Cl_03_pred_Scaled_Petals'], cmap='rainbow' )
axs[1].set_title("3 Clusters on Petal Attributes")
axs[1].set_xlabel('Petal length')
axs[1].set_ylabel('Petal width')

plt.tight_layout()
plt.show()
```

Further clarifications
In fact, if you read about it, the original dataset has 3 sub-species of the Iris flower. Therefore, the number of clusters is 3.

This shows us that:

the Eblow method is imperfect (we might have opted for 2 or even 4)
k-means is very useful in moments where we already know the number of clusters - in this case: 3
biology cannot be always quantified with k-means!
